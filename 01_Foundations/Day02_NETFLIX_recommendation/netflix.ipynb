# **MODELLING**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def weight(x):
  director = (str(x['director']) + ' ') * 9
  actors = (str(x['cast']) + ' ') * 8
  genres = (str(x['listed_in']) + ' ') * 8
  country = (str(x['country']) + ' ') * 9
  description = (str(x['description']) + ' ') * 6
  show = (str(x['type']) + ' ') * 3
  rate = (str(x['rating']) + ' ') * 5
  time = str(x['duration'])



  return director + actors + genres + country + description + show+ rate +time

data['weight'] = data.apply(weight, axis=1)
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(data['weight'])
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

```

```python
def check_my_custom_model(movie_title):
    if movie_title not in data['title'].values:
        print(f" Error: '{movie_title}' not found.")
        return

    idx = data[data['title'] == movie_title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_matches = sim_scores[1:6]

    print(f"--- TESTING WEIGHTS FOR: '{movie_title}' ---")
    print(f"Original Country: {data.loc[idx, 'country']}")
    print(f"Original Director: {data.loc[idx, 'director']}")
    print("-" * 50)

    for i, score in top_matches:
        match_title = data['title'].iloc[i]
        match_country = data['country'].iloc[i]
        match_director = data['director'].iloc[i]


        print(f"Match: {match_title}")
        print(f"   Score: {round(score, 3)}")
        print("-" * 50)

check_my_custom_model("Mha")
check_my_custom_model('Jeans')
check_my_custom_model("Zubaan")
check_my_custom_model('Kota Factory')
check_my_custom_model('Jaws')
```

# **IMPROVSING THE MODELS**

**IMPORTING MODULES**

```python
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

**Loading data**

```python
data = pd.read_csv("/content/netflix_titles.csv")
broken_rows = data['duration'].isnull()
data.loc[broken_rows, 'duration'] = data.loc[broken_rows, 'rating']
data.loc[broken_rows, 'rating'] = ''
data = data.fillna('')
```

**Cleaning data**

*country and text cleaning*

```python
def clean_data(x):
    return str.lower(x.replace(" ", ""))


cols = ['director', 'cast', 'country', 'listed_in', 'rating', 'type']
for col in cols:
    data[col] = data[col].apply(clean_data)
data['description_cleaned'] = data['description'].apply(lambda x: str.lower(x))

data['title_id'] = data['title'].apply(clean_data)
```

**FEATURE ENGINEERING**

*STRICT FILTERING*

```python
def get_rating_group(rating):
    rating = str(rating).lower()
    if rating in ['tv-y', 'tv-g', 'g', 'tv-y7', 'tv-y7-fv', 'pg', 'tv-pg']:
        return 'family'
    elif rating in ['pg-13', 'tv-14']:
        return 'teens'
    elif rating in ['r', 'tv-ma', 'nc-17']:
        return 'adults'
    else:
        return 'unrated'

data['rating_group'] = data['rating'].apply(get_rating_group)
```

**DEFINE SOUP**

```python
data['primary_country'] = data['country'].apply(lambda x: x.split(',')[0].strip().lower() if x else '')
data['title_clean'] = data['title'].apply(lambda x: str.lower(x).strip())
data['description_clean'] = data['description'].apply(lambda x: str.lower(x).strip())

```

```python
# 1. Create the Soup (Director + Cast + Description)
# This combines all important info into one string for every movie
def create_soup(x):
    director = (str(x['director']).lower().replace(" ", "") + ' ') * 8
    cast = (str(x['cast']).lower().replace(" ", "") + ' ') * 6
    desc = x['description_clean']
    return director + cast + desc

data['soup'] = data.apply(create_soup, axis=1)

# 2. Create the Similarity Matrix
# We calculate the math here and save it to 'cosine_sim'
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(data['soup'])

# IMPORTANT: We name this 'cosine_sim' so your function can find it!
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

**Search FUNCTION**

```python
def find_movie_title(user_input):
    user_input = str(user_input).lower().strip()

    if user_input in data['title_clean'].values:
        return data.loc[data['title_clean'] == user_input, 'title'].iloc[0]

    matches = difflib.get_close_matches(user_input, data['title'].tolist(), n=1, cutoff=0.5)
    if matches: return matches[0]

    for title in data['title']:
        if user_input in title.lower():
            return title
    return None
```

#**ENGINE**

```python
def get_recommendations(user_input):
    title = find_movie_title(user_input)
    if title is None:
        print(f"Movie '{user_input}' not found.")
        return

    idx = data[data['title'] == title].index[0]
    target = data.iloc[idx]
    print(f"--- TESTING WEIGHTS FOR: '{title}' ---")
    print(f"Original Country: {target['primary_country']}")
    print(f"Original Director: {target['director']}")
    print("-" * 50)

    candidates = []
    # Use the SOUP similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))

    for i, score in sim_scores:
        if i == idx: continue

        cand = data.iloc[i]
        t_clean = target['title_clean']
        c_clean = cand['title_clean']
        is_franchise = (t_clean in c_clean) or (c_clean in t_clean)

        # Franchise Check (with safety for short titles)
        shortest = t_clean if len(t_clean) < len(c_clean) else c_clean
        if len(shortest) < 4:
            is_franchise = c_clean.startswith(t_clean) or t_clean.startswith(c_clean)


        # Strict Filters (Use rating_group, NOT rating)
        passes_filters = False
        if (cand['rating_group'] == target['rating_group']) and \
           (cand['primary_country'] == target['primary_country']) and \
           (cand['type'] == target['type']):
            passes_filters = True

        if is_franchise or passes_filters:
            final_score = score + (10.0 if is_franchise else 0.0)
            candidates.append((cand['title'], final_score, cand['rating'], cand['primary_country'], cand['type']))

    candidates = sorted(candidates, key=lambda x: x[1], reverse=True)
    top_10 = candidates[:10]

    for c_title, c_score in top_10:
        print(f"Match: {c_title}")
        print(f"   Score: {c_score:.3f}")
        print("-" * 50)
```

```python
get_recommendations("JAWS")
```
